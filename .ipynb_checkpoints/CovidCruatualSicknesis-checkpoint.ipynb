```python
from sklearn.datasets import load_boston
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  
from scipy import stats # При работе со статистикой
```

```python
df = pd.read_excel('Готовая таблица_Симушев_2.xlsx')
# df["Ctreating. 1"] = df["Ctreating. 1"].str.replace(r"[^\d\.]", "1", regex=True)
df.head(10000)
# df["Ctreating. 1"] = df["Ctreating. 1"][1].astype(float).fillna(0)
```

```python
def zavisimaya(row):
    try:
        return ((row['Age'])*16.  + (row['Temperature.'])*6 + (row['SatO2'])*8 
                + (row['Neutral.1'])*13  + row['Ctreating. 1']*16 + row['SRB']*26+ row['fibrinogen']*15 )*0.01
    except:
        return 0
# fd = df.drop(index = [0])
```

```python
df['zavisimaya'] = df.apply(zavisimaya, axis = 1)
df.head(1000)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
df.drop(index = [0])
```

```python
fd = df.drop(labels = [0], axis = 0)
fd.head(1000)
```

```python
X = fd[['Age', 'Temperature.', 'SatO2',
       'Neutral.1', 'SRB', 'Ctreating. 1', 'fibrinogen']]
y = fd['zavisimaya']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=200)
```

```python
X_const = sm.add_constant(X_train) # техническая особенность библиотек, надо руками добавить константу
```

```python
X_const.shape
```

```python
y
```

```python
X
```

```python
model = sm.OLS(y_train, X_const.astype(float))
result = model.fit()
print(result.summary())
```

```python
# df['X_valuation'] = 
def x_valuation(row):
    if row <= 40:
        return 0
    elif 41 <= row <=63:
        return 1
    else:
        return 2
```

```python
df['x_valuation'] = df['Критерий Х'].apply(x_valuation)
df.head(1000)
```

```python
def clinic_x_val(row):
    if row == 'Т':
        return 2
    if row == 'Л':
        return 0
    else:
        return 1
```

```python
df['x_clinic'] = df['Клин_Оценка'].apply(clinic_x_val)
```
